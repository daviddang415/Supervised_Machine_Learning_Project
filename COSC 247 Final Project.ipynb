                                                The NBA’s Next Superstar

Context: 
    
    The National Basketball Association (NBA) hosts an annual draft where NBA teams select college/international prospects with the hopes that they will develop into great players. Along with the NBA draft, NBA teams can sign undrafted players to play for them based on their performance in summer leagues. Despite their best efforts to accurately predict the next NBA Superstar, NBA teams often struggle to determine if a player’s game will translate into the NBA. As a result, a prospect’s performance in their first season in the NBA, known as their rookie season, is often a good indicator of their potential as an NBA player. As such, I aim to use machine learning to determine if an NBA player will become a Superstar using their regular season statistics as a rookie. Specifically, I define a Superstar to be a player that has been selected as an All-Star and earned All-NBA honors at least once throughout their career. An All-Star is a player that has been recognized as one of the best players in their respective conferences. An All-NBA player is one of the best players of their position for the regular season. Additionally, by using machine learning on NBA rookies, I aim to discover trends and find particular statistics that help identify future NBA Superstars. Along with NBA Superstars, I wish to determine which NBA players will become future All-Stars as there are many excellent players that deserve recognition for their skills despite not being a Superstar caliber player. 


Procedure:

    Firstly, I created the NBA rookie performance dataset to help output correlation matrices, feature graphs, and classification model predictions. With the dataset, I collected statistics from the regular season of NBA rookies from the years 2009-2019 using the basketball-reference website (https://www.basketball-reference.com/). The statistics that I decided to use to train the classification models were age, the number of games played, field goal percentage, three-point percentage, free throw percentage, minutes per game, points per game, total rebounds per game, and assists per game. An obstacle I encountered creating the dataset, however, was that some cells from the shooting statistics were empty, which indicated that a player did not attempt a specific shot. As a result, I denoted a negative one in the cell to represent that a player did not shoot either a field goal, three-pointer, or free-throw in a season.
    
    Also, I researched online the NBA rookies from the years 2009 to 2019 that managed to become Superstars based on the criteria that they became an All-Star and earned All-NBA honors. Before inputting the data on Superstar players, I decided to research which rookies only managed to become an All-Star or earn All-NBA honors. As a result, in a separate column, I input one if a rookie had been selected as an All-Star in his career and zero if he was a non-All-Star. Interestingly, I noticed that every player that earned All-NBA honors also became an All-Star in their career. As a result, for a player that earned All-NBA honors, I wrote a one, two, or three according to the highest honors they received in their career in a different column. After that, I created another column to identify NBA rookies that became Superstars. I inputted a one for a Superstar player and a zero for a non-Superstar player. Additionally, I created a few different datasets that consisted of various Superstars and All-Stars to use them for the correlation matrices. After organizing the data, I printed out the entire dataset, the Superstars, All-Stars, and the non-Superstars.
    
    Before utilizing the classification models, I determined which features correlated with one another that potentially impacted the chance of a player to become a Superstar with a Matplotlib correlation matrix that used a heatmap. I believed that there was a positive correlation between points per game and field goal percentage that impacted a player’s chances to become a Superstar. I also believed that there was a negative association between total rebounds per game and free throw percentage. For the feature pairs that had the highest correlation according to the heatmap, I graphed them on a scatter plot to observe their linear association. Also, to better understand NBA rookies that became Superstars, I decided to create correlation matrices for each All-NBA team to learn more about the different levels between NBA superstars. I believe that there is a correlation between age and points per game for All-NBA First Team players that helps explain the reason they are the best players in the league. Once again, I graphed the feature pairs from each correlation matrix to observe various correlations between the NBA Superstars.
    
    Although many NBA rookies that develop throughout their career only become All-Stars, I still think these players are great due to the accomplishment of earning such a title. As such, I decided to graph a heatmap correlation matrix to determine the strength of association between the features from NBA rookies that became All-Stars. I believe that there is a positive correlation between points per game and age that impacts a player's chances to become an All-Star. By observing such correlations from NBA Superstars and All-Stars, I can better understand which statistics are important in determining which NBA rookies will become future stars in the league.
    
    For the classification models, I used the perceptron, stochastic gradient descent, and random forest classification models from sklearn to predict which rookies would become future Superstars. My goal is that the classification models could predict the testing set with 93% accuracy. With the dataset, I split the data into training and testing sets. Also, I standardized the dataset according to the training set to increase the performance of the classification models. Learning from the project prototype assignment, I used the random state hyperparameter in each classifier to increase the model's accuracy. For the perceptron, I used a learning rate of 0.01 with a maximum number of iterations of 10,000 and set the random state to true. In terms of the stochastic gradient descent, I used a learning rate of 0.01 with a maximum number of iterations of 10,000 along with a hinge loss and l2 regularization with the random state set to true. For the random forest classifier, I used its default parameters, except changed the random state to true, to classify the data. After setting the hyperparameters, I trained the models using the standardized training set. I then determined the accuracy of each classification model in predicting the testing set. Also, I created a feature importance graph for each classification model to determine which features each model deemed the most useful in predicting which players had the potential to become Superstars. In my opinion, I believe that points per game would be the best feature in determining if an NBA rookie would become a Superstar. 
    
    Additionally, I used another set of the perceptron, stochastic gradient descent, and random forest classification models with the same hyperparameters to learn more about NBA All-Stars. I wanted to explore the statistics that help determine if a certain rookie would become an All-Star. As such, I created feature graphs for each classifier to determine the importance of each statistic for an NBA All-Star. I believe that field goal percentage is the most important statistic for an NBA rookie to become an All-Star. Also, I wish for the accuracy of the classification models to exceed 93%. For curiosity, I plan on using the most accurate classification model from the three machine learning models from the NBA Superstar data and NBA All-Star data to predict unseen data consisting of NBA rookie performances from other regular seasons.

```python
#Importing dataset, splitting the features and targets, and dividing the dataset into training and testing sets (one data set based on NBA All-Stars and another on NBA Superstar)
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

df = pd.read_csv('2009-2019 Rookie Data Set.csv')

dataList = df.to_numpy()

data = dataList[:, 1:10]
data = data.astype('float')

tempTarget = dataList[:, 10]
tempTarget = tempTarget.astype('int')

target = dataList[:, 12]
target = target.astype('int')

temp_X, temp_y = data, tempTarget

temp_X_train, temp_X_test, temp_y_train, temp_y_test =\
        train_test_split(temp_X, temp_y, test_size=0.3, random_state=1, stratify=temp_y)

X, y = data, target

X_train, X_test, y_train, y_test =\
        train_test_split(X, y, test_size=0.3, random_state=1, stratify=y)
```

```python
#Standardizing both the training set and testing set according to the training set
from sklearn.preprocessing import StandardScaler

temp_sc = StandardScaler()
temp_sc.fit(temp_X_train)

temp_X_train_std = temp_sc.transform(temp_X_train)
temp_X_test_std = temp_sc.transform(temp_X_test)

temp_X_combined_std = np.vstack((temp_X_train_std, temp_X_test_std))
temp_y_combined = np.hstack((temp_y_train, temp_y_test))
```

```python
#Standardizing both the training set and testing set according to the training set
sc = StandardScaler()
sc.fit(X_train)

X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)

X_combined_std = np.vstack((X_train_std, X_test_std))
y_combined = np.hstack((y_train, y_test))
```
